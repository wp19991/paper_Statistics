```python
import requests
import time
import csv
import json
```

```python
def from_cff_index_ids_list_to_name(li):
    name_list = []
    for i in li:
        url = "https://www.acemap.info/conference-series/{}".format(i.split("_")[0])
        name_temp = requests.get(url).text.split("<title>")[1].split("</title>")[0]
        if name_temp == "Server error":
            url = "https://www.acemap.info/journal/{}".format(i.split("_")[0])
            name_temp = requests.get(url).text.split("<title>")[1].split("</title>")[0]
        name_list.append(name_temp)
    return name_list
```

```python
def from_id_year_cff_to_count(affiliation_id,year,cff_id):
    affiliation_list_url = "https://www.acemap.info/api/v2/aceranking/affiliation_list"
    affiliation_list_post_data = {
        "index": "CCF",
        # 顶级会议的类别
        "venue_ids": cff_id,
        "area": "All",
        "start_year": year, # 开始年份
        "end_year": year, # 结束年份
        "first_author": "0",
        "order_by": "2",#排序方式 2 - 论文被引用次数
        "affiliation_id": affiliation_id # 公司编号
    }
    author_list_url_text = requests.post(affiliation_list_url,data=affiliation_list_post_data).text
    #print(author_list_url_text)
    #print(author_list_url_text.split("\"count\":")[1].split(",")[0])
    if len(author_list_url_text)!=2:
        temp_num_str = author_list_url_text.split("\"count\":")[1].split(",")[0]
    else:
        temp_num_str = "0"
    
    return temp_num_str
```

```python
affiliation_id_list = [
    "2100597028", # 百度
    "2102136657", # 腾讯
    "2101566148", # 阿里
]
affiliation_id_name = [
    "百度",
    "腾讯",
    "阿里"
]

cff_index_ids_list = [
    "2122770072_0","2122280177_0","2123670350_0","2120665855_0",
    "2122547259_0","2122362464_0","2124815797_0","2114134642_0",
    "2112195503_0","2113800408_0","2110409127_0"
]

cff_index_ids_name_list =['National Conference on Artificial Intelligence - Acemap',
                          'Computer Vision and Pattern Recognition - Acemap',
                          'Meeting of the Association for Computational Linguistics - Acemap',
                          'International Conference on Computer Vision - Acemap',
                          'International Joint Conference on Artificial Intelligence - Acemap',
                          'International Conference on Machine Learning - Acemap',
                          'Neural Information Processing Systems - Acemap',
                          'IEEE Transactions on Pattern Analysis and Machine Intelligence - Acemap',
                          'International Journal of Computer Vision - Acemap',
                          'Artificial Intelligence - Acemap',
                          'Journal of Machine Learning Research - Acemap']

searh_year_list = ["2017","2018","2019","2020"]
```

```python
year_name_cffname_cffid_count_list=[]

#每个年份
for year in searh_year_list:
    
    # 每个公司
    for affiliation_id,name in zip(affiliation_id_list,affiliation_id_name):
        
        #论文的标签
        for cff_id,cff_name in zip(cff_index_ids_list,cff_index_ids_name_list):
            time.sleep(0.5)
            temp_year = str(year)
            temp_name = name
            temp_cffname = cff_name
            temp_cffid = cff_id
            #print(temp_year,temp_name,temp_cffid)
            
            temp_count = from_id_year_cff_to_count(affiliation_id,temp_year,cff_id)
            temp_list = [temp_year,temp_name,temp_cffname,temp_cffid,temp_count]
            print(temp_year,temp_name,temp_count,temp_cffname)
            year_name_cffname_cffid_count_list.append(temp_list)
            
```

```python
headers = ['year','group_name','hiuyi_name','cffid','count']

with open("cff_paper.csv","w",encoding="utf-8",newline='')as f:
    f_csv = csv.writer(f)
    f_csv.writerow(headers)
    f_csv.writerows(year_name_cffname_cffid_count_list)
```
